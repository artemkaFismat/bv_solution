```python
file = open('/content/numbers.txt', 'r')
content = file.read().split(' ')
file.close()
m = 0
n = 0
for num in content:
  c = 0
  for i in content:
    if num == i:
      c += 1
  if c > m:
    m = c
    n = num
print(n)
```

```python
# 22-24
import math
import numpy

def sigmoida(x):
  return 1 / (1 + math.exp(-x))

z = 1 + 7 * -0.2 + 7 * 0.3
print(sigmoida(z))
print(numpy.sign(sigmoida(z) - 0.5))
```

```python
from pandas import *

data = read_csv("/content/actions.csv")

d = data['date'].tolist()
a = data['action'].tolist()

c = 0
for i in range(len(d)):
  if d[i].split('.')[1] == '03' and a[i] == 'click':
    c += 1
print(c)
```

```python
from pandas import *
from collections import Counter

data = read_csv("/content/actions.csv")

d = data['date'].tolist()
ip = data['ip'].tolist()

c = []
for i in range(len(d)):
  if d[i].split('.')[1] == '03':
    c.append(ip[i])

print(len(Counter(c)))
```

```python
file = open("/content/_graph (1).txt", "r")
m = 0
while file.readline() != '':
  content = file.readline()
  c = 0
  for i in content.split(','):
    if i == '1':
      c += 1
  if c > m:
    m = c
print(m)
file.close()
```
